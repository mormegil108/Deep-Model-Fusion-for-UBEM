### Create building-specific input data

```python
save_directory = r"C:\Users\sb013698\Desktop\github\Deep Model Fusion for UBEM\Create Input Data\Input Data"
```

```python
for b in building_dict.keys():

    building_name = b
    building = building_dict[b]

    # 1) Define archtype and its index
    archetype = building[0]
    arc_index = archetype_index[archetype]
    
    # Set details at building level (static)
    actual_gfa = building[1]
    rhino_gfa = building[2]
    wwr = building[3]
    n_floors = building[4]
    height = building[5]
    volume = building[5]

    # 2) Set details at archetype level
    # Schedules (time-dependent)
    heating_schedule = generated_schedules.loc[:, year_schedules_dict[archetype_dict["Heating Schedule"][arc_index]]] # numpy array: (8760,)
    cooling_schedule = generated_schedules.loc[:, year_schedules_dict[archetype_dict["Cooling Schedule"][arc_index]]] # numpy array: (8760,)
    dhw_schedule = generated_schedules.loc[:, year_schedules_dict[archetype_dict["DHW Schedule"][arc_index]]] # numpy array: (8760,)
    equipment_schedule = generated_schedules.loc[:, year_schedules_dict[archetype_dict["Equipment Schedule"][arc_index]]] # numpy array: (8760,)
    lighting_schedule = generated_schedules.loc[:, year_schedules_dict[archetype_dict["Lighting Schedule"][arc_index]]] # numpy array: (8760,)
    occupancy_schedule = generated_schedules.loc[:, year_schedules_dict[archetype_dict["Occupancy Schedule"][arc_index]]] # numpy array: (8760,)
    nat_vent_schedule = generated_schedules.loc[:, year_schedules_dict[archetype_dict["Natural Ventilation Schedule"][arc_index]]] # numpy array: (8760,)
    mech_vent_schedule = generated_schedules.loc[:, year_schedules_dict[archetype_dict["Mechanical Ventilation Schedule"][arc_index]]] # numpy array: (8760,)

    # 3) Static features from archetype_dict and building_dict
    dhw_flow_rate = archetype_dict["DHW Flow Rate (m3/h/m2)"][arc_index]
    epd = archetype_dict["EPD (W/m2)"][arc_index]
    lpd = archetype_dict["LPD (W/m2)"][arc_index]
    illuminance = archetype_dict["IlluminanceTarget (lux)"][arc_index]
    occupant_density = archetype_dict["Occupant Density (people/m2)"][arc_index]
    infiltration = archetype_dict["Infiltration"][arc_index]
    heating_setpoint = archetype_dict["Heating Setpoint"][arc_index]
    cooling_setpoint = archetype_dict["Cooling Setpoint"][arc_index]
    wall_u = archetype_dict["wall u-value"][arc_index]
    roof_u = archetype_dict["roof u-value"][arc_index]
    ground_u = archetype_dict["ground u-value"][arc_index]
    window_u = archetype_dict["window u-value"][arc_index]
    is_cooling_on = archetype_dict["IsCoolingOn"][arc_index]
    is_mech_vent_on = archetype_dict["IsMechVentOn"][arc_index]

    # 4) Target: Hourly energy consumption data
    if building_name in target_data["Building Name"].value_counts().index.to_list():
        target_energy = target_data[target_data["Building Name"]==building_name].iloc[:, 1::].values.reshape(-1,) # numpy array: (8760,)
    else:
        print(f"Error: {building_name} is not available in the metered data!")

    # 5) Define the weather data: ['Dry Bulb Temperature {C}', 'Dew Point Temperature {C}', 'Relative Humidity {%}', 'Wind Direction (deg)', 'Wind Speed (m/s)', 'Hourly Total Global Solar Radiation (watt/m2)', 'Liquid Precipitation (mm)', 'Sun Exposure Intensity (cal/cm2)']
    hourly_weather = weather_data.copy() # pandas dataframe: (8760, 8)

    # 6) Add data times: "Month", "Day", "Weekday", "Hours"
    data_times = calendar.copy() # pandas dataframe: (8760, 4)

    # 7) Calculate the new features for Hourly DHW, Hourly Equipment, Hourly Lighting, Hourly Occupancy
    hourly_dhw = dhw_schedule * dhw_flow_rate
    hourly_equipment = equipment_schedule * epd
    hourly_lighting = lighting_schedule * lpd
    hourly_occupancy = occupancy_schedule * occupant_density

    # 8) Calcualte the difference between outside temperature and heating/cooling setpoints
    heating_delta_t = list()
    cooling_delta_t = list()
    for i in range(heating_schedule.shape[0]):
        if heating_schedule[i] == 0:
            delta = 0
            heating_delta_t.append(delta)
        else:
            delta = (heating_setpoint - hourly_weather['Dry Bulb Temperature {C}'][i]) * heating_schedule[i]
            heating_delta_t.append(abs(delta))
    
    for i in range(cooling_schedule.shape[0]):
        if cooling_schedule[i] == 0:
            delta = 0
            cooling_delta_t.append(delta)
        else:
            delta = (hourly_weather['Dry Bulb Temperature {C}'][i] - cooling_setpoint) * cooling_schedule[i]
            cooling_delta_t.append(abs(delta))

    # Order: 'Archetype', 'Actual GFA (m2)', 'Rhino GFA (m2)', 'WWR', 'Floor Count', 'Height', 'Volume'
    # 8) Prepare the final input dataframe
    input_data = pd.DataFrame({
        'Building Name': building_name,
        'Archetype': building_dict[building_name][0],
        'Month': data_times['Month'],
        'Day': data_times['Day'],
        'Weekday': data_times['WeekDay'],
        'Hours': data_times['Hours'],
        'Dry Bulb Temperature {C}': hourly_weather['Dry Bulb Temperature {C}'],
        'Dew Point Temperature {C}': hourly_weather['Dew Point Temperature {C}'],
        'Relative Humidity {%}': hourly_weather['Relative Humidity {%}'],
        'Wind Direction (deg)': hourly_weather['Wind Direction (deg)'],
        'Wind Speed (m/s)': hourly_weather['Wind Speed (m/s)'],
        'Hourly Total Global Solar Radiation (watt/m2)': hourly_weather['Hourly Total Global Solar Radiation (watt/m2)'],
        'Liquid Precipitation (mm)': hourly_weather['Liquid Precipitation (mm)'],
        'Sun Exposure Intensity (cal/cm2)': hourly_weather['Sun Exposure Intensity (cal/cm2)'],
        'Heating Schedule': heating_schedule,
        'Cooling Schedule': cooling_schedule,
        'DHW Schedule': dhw_schedule,
        'Equipment Schedule': equipment_schedule,
        'Lighting Schedule': lighting_schedule,
        'Occupant Schedule': occupancy_schedule,
        'Natural Ventilation Schedule': nat_vent_schedule,
        'Mechanical Ventilation Schedule': mech_vent_schedule,
        'DHW Flow Rate (m3/h/m2)': [dhw_flow_rate] * 8760,
        'EPD (W/m2)': [epd] * 8760,
        'LPD (W/m2)': [lpd] * 8760,
        'Hourly DHW': hourly_dhw,
        'Hourly Equipment': hourly_equipment,
        'Hourly Lighting': hourly_lighting,
        'Hourly Occupancy': hourly_occupancy,
        'Illuminance (lux)': [illuminance] * 8760,
        'Occupant Density (people/m2)': [occupant_density] * 8760,
        'Infiltration (ACH)': [infiltration] * 8760,
        'Heating Setpoint (celsius)': [heating_setpoint] * 8760,
        'Cooling Setpoint (celsius)': [cooling_setpoint] * 8760,
        'Heating Delta-T': heating_delta_t,
        'Cooling Delta-T': cooling_delta_t,
        'Wall-U': [wall_u] * 8760,
        'Roof-U': [roof_u] * 8760,
        'Ground-U': [ground_u] * 8760,
        'Window-U': [window_u] * 8760,
        'IsCoolingOn': [is_cooling_on] * 8760,
        'IsMechVentOn': [is_mech_vent_on] * 8760,
        'WWR': building_dict[building_name][3],
        'Actual GFA (m2)': building_dict[building_name][1],
        'Rhino GFA (m2)': building_dict[building_name][2],
        'Floor Count': building_dict[building_name][4], 
        'Height': building_dict[building_name][5], 
        'Volume': building_dict[building_name][6],
        'Target Hourly Energy (kWh)': target_energy,
    })

    # 8) Save the input data as a CSV file for each building
    input_name = f"{file_name_dict[building_name]}.csv"
    input_data.to_csv(os.path.join(save_directory, input_name), index=False)
    print(f"Data for {building_name} was saved as >> {input_name}")
```
